## Total matches played in each Season

```python
plt.figure(figsize = (16,10))
sns.countplot('season',data=matches,palette="winter")
```

## Most used venue

```python
ls=matches['venue'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(15,10))
temp=sns.barplot(ls.index, ls.values, alpha=0.9)
plt.title('Stadium with most matches')
plt.ylabel('Count')
plt.xlabel('Stadiums')
plt.xticks(rotation=90)
plt.show()
```

we can see that eden Gardens is the Most loved venue , followed by wankhede and chinnaswamy.

## Matches Per team

```python
matches_per_team1=matches.groupby('team2')['team2'].count()
matches_per_team2=matches.groupby('team1')['team1'].count()

matches_per_team = matches_per_team1 + matches_per_team2
matches_per_season=matches.groupby('season')['season'].count()
matches_per_team
```

## Numbers of matches won by team

```python
plt.figure(figsize = (18,10))
sns.countplot(x='winner',data=matches)
plt.title("Numbers of matches won by team ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.ylabel("No of wins",fontsize=15)
plt.show()
```

## Victory vs Venue

```python
venue_wins=matches.groupby(['venue','winner']).apply(lambda x:x['winner'].count()).reset_index(name='Match wins')
venue_wins_pvt=pd.pivot(venue_wins,values='Match wins',index='venue',columns='winner')
venue_wins_pvt.replace(np.NaN,0) #pivot plot
plt.figure(figsize=(20,20))
heat=sns.heatmap(venue_wins_pvt,annot=True)
plt.xlabel('Teams',size=10)
plt.ylabel('Venues',size=10)
plt.title('Teams wins at different venues',size=20)
```

### Numbers of matches won by batting and bowling first 

```python
matches['win_by']=np.where(matches['win_by_runs']>0,'Bat first','Bowl first')
```

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue=matches['win_by'],data=matches,palette='cool')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()

Win=matches.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
plt.figure(figsize = (5,6))
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

## graph on Numbers of matches won by Toss result

```python
# we will plot graph on Numbers of matches won by Toss result
plt.figure(figsize = (18,10))
sns.countplot('winner',hue='toss_decision',data=matches,palette='cool')
plt.title("Numbers of matches won by Toss result ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.ylabel("No of wins",fontsize=15)
plt.show()
plt.show()

Toss=matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=20)
plt.axis('equal')
plt.show()
```

## winning the toss and winning the match

```python
matches['win_by']=np.where(matches['winner']==matches['toss_winner'],'Bat first','Bowl first')

plt.figure(figsize = (18,10))
sns.countplot('season',hue=matches['win_by'],data=matches)
plt.title(" matches won along with toss ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()

Win=matches.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
plt.figure(figsize = (5,6))
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

## season wise winners

```python
final_matches=matches.drop_duplicates(subset=['season'], keep='last')
final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

## Final match winner probablity

```python
matches['win_by']=np.where(matches['win_by_runs']>0,'Bat first','Bowl first')

match = final_matches.win_by.value_counts()
labels=np.array(Toss.index)
sizes = match.values
colors = ['blue', 'red']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

# Players analysis

## Most Runs in IPL

```python
plt.figure(figsize=(15,10))
deliv.groupby(['batsman']).batsman_runs.sum().sort_values(ascending=False)[:20].plot(kind='bar',color=['red', 'blue', 'purple'])
plt.ylabel('Count')
plt.xlabel('Batsmans')
plt.title('Batsman with Most runs')
plt.xticks(rotation=90)
plt.show()
```

## best 3 players from each season

```python
plt.figure(figsize=[38,25])
plt.title('Top 3 Player of Season',fontsize=12)
season1=2008;
for i in range(1,13):
    plt.subplot(3,4,i)
    plt.title(season1)
    matches[matches.season == season1 ].player_of_match.value_counts()[:3].plot(kind='bar', color=['red', 'blue', 'purple'])
    plt.xticks(rotation=90)
    season1 = season1+1
plt.show;
```

## Most Valued Players

```python
plt.figure(figsize = (15,10))
top_players = matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,22])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="hsv");
plt.show()
```
