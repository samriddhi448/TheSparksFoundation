### Statewise analytics of Profit, Discount, Sales

````python
df['State'].value_counts().head(10)
````

````python
df_states = df.groupby(['State'])[['Sales', 'Discount' , 'Profit']].mean()
df_states.head(10)
````

# Profit analysis Statewise

````python
df5 = df_states.sort_values('Profit')

df5['Profit'].plot(kind = 'bar', figsize = (15,5))
plt.xlabel('States')
plt.ylabel('Profit per State')
plt.title('Profit Analysis Statewise', fontsize = 20)
plt.legend()
````

**From the above plot, we can make out the highest profit per state as well as the lowest**

**Highest Profit State**
1.) Vermount

**Lowest Profit State**
2.) Ohio

# Sales analysis Statewise

````python
df5['Sales'].plot(kind = 'pie', figsize = (20,20), shadow = True, autopct='%1.1f%%',
                        startangle=90)
plt.title('Sales Analysis Statewise', fontsize = 20)
````

**Wyoming has the highest sales in comparison to other states in the country**

**South Dakota has the lowest of all the other states in the country**

# Discount analysis Statewise

````python
df5['Discount'].plot(kind = 'bar', figsize = (15, 5))
                    
plt.title('Discount Analysis Statewise', fontsize = 20)
````

**Illinois having the highest discount Statewise**

## Sub Category wise analysis of profit, sales and discount.

````python
df_sub_category = df.groupby('Sub-Category')[['Profit', 'Sales', 'Discount']].mean()
df_sub_category
````

````python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
````

**Copiers and Machines have highest sales than any other sub category, wheras the least is fasteners**

````python

plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
````

**Binders and machines have the highest discount of all the other sub categories**

````python
df_sub_category.sort_values('Profit')[['Sales', 'Profit']].plot(kind = 'bar', figsize = (15, 5))
````

**Copiers have the highest Sales and Profit apart from other sub categories.**
